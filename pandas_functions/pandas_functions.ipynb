Задание 1**
Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class

```python
import pandas as pd
```

```python
data = pd.read_csv('D:\\Python_homework\\pyda_homeworks\\pandas_functions\\ml-latest-small\\ratings.csv')
```

```python
data.head()
```

```python
def movie_classification(row):
    if row <= 2.0:
        return 'низкий рейтинг'
    elif 2 < row <=4:
        return 'средний рейтинг'
    elif row >= 4.5:
        return 'высокий рейтинг'
    
```

```python
data['movie_classification'] = data['rating'].apply(movie_classification)
data
```

Задание 2
Используем файл keywords.csv.

Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определенному региону. Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим ‘undefined’. Результат классификации запишите в отдельный столбец region.

```python
geo_data = pd.read_csv('D:\\Python_homework\\pyda_homeworks\\pandas_functions\\ml-latest-small\\keywords.csv')
```

```python
geo_data.head()
```

```python
def geo_classification(string):
    if 'москва' in string or 'тула' in string or 'ярославль' in string:
        return 'Центр'
    elif 'петербург' in string or 'псков' in string or 'мурманск' in string:
        return 'Северо-Запад'
    elif 'владивосток' in string or 'сахалин' in string or 'хабаровск' in string:
        return 'Дальний Восток'
    else:
        return 'undefined'
```

```python
geo_data['region'] = geo_data['keyword'].apply(geo_classification)
geo_data
```
