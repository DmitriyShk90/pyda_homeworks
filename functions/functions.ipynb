# Домашнее задание к лекции "Функции"

Вам нужно помочь секретарю автоматизировать работу. Для этого нужно написать программу, которая будет на основе хранимых данных исполнять пользовательские команды.

Исходные данные имеют следующую структуру:

1. перечень всех документов
```
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
```
2. перечень полок, на которых хранятся документы (если документ есть в documents, то он обязательно должен быть и в directories)
```
directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}
```

Общие требования к программе:
- код должен быть грамотно декомпозирован (каждая функция отвечает за свою конкретную задачу, дублирующийся функционал переиспользуется, а его код не повторяется);
- в коде отсутствуют глобальные переменные (за исключением **documents** и **directories**);
- пользовательский ввод обрабатывается в цикле **while** до тех пор, пока пользователь явно не завершит программу (вводом команды "**q**").

````python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
````

````python
directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}
````

````python
# Формула поиска владельца документа по его номеру
def doc_num(document_number):
    for i in documents:
        if i['number'] == document_number:
            return "Владелец документа: ", i['name']
    return "Документ не найден в базе"
        
````

````python
doc_num('10006')
    
````

````python
# Формула поиска полки по номеру документа
def shelf_num(document_number):
    for k,v in directories.items():
        if document_number in v:
            return "Документ хранится на полке: ",k
    return "Документ не найден в базе"
````

````python
shelf_num('10006')
````

````python
# Формула для отображения полной информации по всем документам
def info_():
    for s in directories:
        for n in directories[s]:
            f = next(e for e in documents if e['number']==n)
            print(f'№:{n}, тип:{f["type"]}, владелец:{f["name"]}, полка хранения: {s}')
            
        
````

````python
info_()
````

````python
# Формула для добавления полки
def add_shelf(number_):
    if str(number_) not in directories.keys():
        directories[str(number_)] = []
        print("Полка добавлена. Текущий перечень полок: ", list(directories.keys()))
    else:
        print('Такая полка уже существует. Текущий перечень полок: ', list(directories.keys()))
    
````

````python
# Формула для удаления полки
def delete_shelf(number_):
    number_1 = str(number_)
    if number_1 not in directories.keys():
        print("Такой полки не существует. Текущий перечень полок: ", list(directories.keys()))
    elif len(directories[number_1]) > 0:
        print("На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: ", list(directories.keys()))
    elif len(directories[number_1]) == 0 and number_1 in directories.keys():
        del directories[number_1]
        print("Полка удалена. Текущий перечень полок: ", list(directories.keys()))
        
````

````python
def main(documents, directories):
    while True:
        user_input = input("Введите команду: ")
        if user_input == 'p':
            document_number = input("Введите номер документа: ")
            print(doc_num(document_number))
        elif user_input == 's':
            document_number = input("Введите номер документа: ")
            print(shelf_num(document_number))
        elif user_input == 'l':
            print(info_())
        elif user_input == 'ads':
            number_ = input("Введите номер полки: ")
            print(add_shelf(number_))
        elif user_input == 'ds':
            number_ = input("Введите номер полки: ")
            print(delete_shelf(number_))
        elif user_input == 'q':
            break    
            
            
            
        
````

````python
main(documents, directories)
````
