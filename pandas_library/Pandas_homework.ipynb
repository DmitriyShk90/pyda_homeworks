Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
power = pd.read_csv('D:\\Python_homework\\pyda_homeworks\\pandas_library\\power.csv')
```

```python
power.head()
```

```python
filtered_countries = power[(power.country == 'Estonia')|(power.country == 'Lithuania')|(power.country == 'Latvia')]
```

```python
filtered_countries_years = filtered_countries[(filtered_countries['year']>=2005) & (filtered_countries['year']<=2010)]
```

```python
filtered_countries_years_categories = filtered_countries_years.query('category == 4 | category == 12 | category == 21')
```

```python
power_consumption = filtered_countries_years_categories.query('quantity > 0')
```

```python
power_consumption['quantity'].sum()
```
